```python
import pandas as pd
import sys
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive

# Mount Drive and set path
drive.mount('/content/drive')
drive_root = '/content/drive/MyDrive'

if drive_root not in sys.path:
    sys.path.append(drive_root)

from src.analysis.visualisation import StylometricVisualizer

# Load data
df = pd.read_csv(f"{drive_root}/aesthetics_project/merged_features.csv")

# Isolate features
exclude_cols = ['doc_id', 'model', 'genre', 'prompt', 'text']
feature_columns = [col for col in df.columns if col not in exclude_cols]

# Instantiate the visualizer (it automatically scales the data)
visualizer = StylometricVisualizer(data=df, feature_cols=feature_columns)
```

```python
# Set a clean theme for the landscape
sns.set_theme(style="white")

print("Generating Aesthetic Landscape by Model...")
print("This shows the 'territory' each LLM's style occupies.")

fig_models = visualizer.plot_aesthetic_landscape(hue_col='model', perplexity=30)
plt.show()
```

```python
print("Generating Aesthetic Landscape by Genre...")
print("This shows the 'territory' each Genre occupies.")
# Using genre as the color hue instead of model
fig_genres = visualizer.plot_aesthetic_landscape(hue_col='genre', perplexity=30)
plt.show()
```
